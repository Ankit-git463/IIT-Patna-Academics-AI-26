```python
import numpy as np 
import pandas as pd 
import cv2 
import matplotlib.pyplot as plt
```

```python
image_path = "102061.png"

image =cv2.imread(image_path)

if image is None: 
    print("Image Not Loaded !! ")
else: 
    print("Image Loaded Successfully !! ")


```

```python
plt.imshow(cv2.cvtColor(image , cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis('off')
```

```python

num = np.random.randint(30 , 256 )
print("The random number is :" ,num)
```

```python
bright_image = image + num

plt.imshow(cv2.cvtColor(bright_image , cv2.COLOR_BGR2RGB))
plt.title("Bright Image")
plt.axis('off')
```

```python
flip_horizontal = cv2.flip(image , 1 )

plt.imshow(cv2.cvtColor(flip_horizontal, cv2.COLOR_BGR2RGB))
plt.title("Horizontal Flipped Image")
plt.axis('off')
```

```python
avg_img = (0.5 * flip_horizontal+ 0.5 * image).astype(np.uint8)

# avg_img = cv2.addWeighted(image , 0.5 , flip_horizontal , 0.5 , 0 )
plt.imshow(cv2.cvtColor(avg_img, cv2.COLOR_BGR2RGB))
plt.title("Average Image")
plt.axis('off')
```

```python
cv2.imwrite("Avg_image.png" , avg_img)
```

```python
rnum = np.random.randint(50 , 200)/100
print("Random Generated Value is : " , rnum)
```

```python
contrast = np.clip(rnum*image ,0 , 255 ).astype(np.uint8) 

plt.imshow(cv2.cvtColor(contrast, cv2.COLOR_BGR2RGB))
plt.title("Contrast Image")
plt.axis('off')
```

```python
r ,g, b = cv2.split(image)
g[:] = 0 

combine = cv2.merge((r,g,b))
plt.imshow(cv2.cvtColor(combine, cv2.COLOR_BGR2RGB))
plt.title("No Green")
plt.axis('off')
```

```python
import os 

folder = "Output Folder"

if not os.path.exists(folder):
    os.makedirs(folder)

outpath = os.path.join( folder , "No Green.png")

print("Path of image is " , outpath)

cv2.imwrite(outpath , combine)


```
