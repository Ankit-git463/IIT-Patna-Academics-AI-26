```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt

import cv2 
```

```python
image_path = "lion.png"

image =cv2.imread(image_path)

if image is None: 
    print("Image Not Loaded !! ")
else: 
    print("Image Loaded Successfully !! ")
```

```python
plt.imshow(cv2.cvtColor(image , cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis('off')
```

```python

gray = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)



```

```python
kernel = np.array([
    [-2 , -1 , 0 ], 
    [-1 , 1 , 1 ], 
    [0 , 1 , 2 ]
])

newimg= cv2.filter2D(image, ddepth=-1, kernel= kernel) 

plt.imshow(cv2.cvtColor(newimg , cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis('off')
```

```python
kernel = np.array([
    [-1 , 0 , 1 ], 
    [-1 , 0 , 1 ], 
    [-1 , 0 , 1 ]
])

newimg= cv2.filter2D(gray , ddepth=-1, kernel= kernel) 

plt.imshow(newimg, cmap='gray')
plt.title("Original Image")
plt.axis('off')
```

```python
kernel = np.array([
    [3 , 0 , -3 ], 
    [10 , 0 , -10 ], 
    [3 , 0 , -3 ]
])

newimg= cv2.filter2D(gray , ddepth=-1, kernel= kernel) 

plt.imshow(newimg, cmap='gray')
plt.title("Original Image")
plt.axis('off')
```
