```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Load the image in grayscale
img = cv2.imread('227092.png', cv2.IMREAD_GRAYSCALE)

```

```python
# Display the original image

plt.title("Original Image")
plt.imshow(img, cmap='gray')
plt.axis('off')


```

```python
# Apply 2D Fourier Transform
f = np.fft.fft2(img)
fshift = np.fft.fftshift(f)  # Shift zero frequency to the center
magnitude_spectrum = 20 * np.log(np.abs(fshift) + 1)  # log scale for visibility


print(magnitude_spectrum.shape)

```

```python

# Display the magnitude spectrum
plt.title("Magnitude Spectrum")
plt.imshow(magnitude_spectrum, cmap='gray')
plt.axis('off')
plt.show()

```

```python
# Inverse Fourier Transform to reconstruct image
f_ishift = np.fft.ifftshift(fshift)
img_reconstructed = np.fft.ifft2(f_ishift)
img_reconstructed = np.abs(img_reconstructed)

```

```python
# Display reconstructed image
plt.title("Reconstructed Image")
plt.imshow(img_reconstructed, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
