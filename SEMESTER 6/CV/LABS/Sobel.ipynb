```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import cv2 
```

```python
image_path = '227092.png'

image = cv2.imread(image_path)

if image is None: 
    print("Image not loaded !!")
else : 

    print("Image Loaded Successfully")
```

```python
def showimg(image , title):
    plt.imshow(cv2.cvtColor(image , cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
showimg(image , "1. Original Image")
```

```python
blur= cv2.GaussianBlur(image , (5 , 5 ) , 3 )

showimg(blur , "2. Blur Image")
```

```python
gray = cv2.cvtColor(blur  , cv2.COLOR_BGR2GRAY)
showimg(gray , "3. Grayscale Image")
```

```python
xkernel = np.array([
    [-1 , 0 , 1 ], 
    [-2 , 0 , 2 ], 
    [-1 , 0 , 1 ]
])
ykernel = np.array([
    [-1 , -2 , -1 ], 
    [0 , 0 , 0  ], 
    [ 1 , 2 , 1 ]
])

print("X operator kernel " , xkernel)
print("Y operator kernel " , ykernel)

```

```python
sobelx = cv2.filter2D(gray , ddepth=cv2.CV_64F , kernel= xkernel) 
sobely = cv2.filter2D(gray , ddepth=cv2.CV_64F , kernel= ykernel) 

plt.subplot(1 , 2 , 1 )

plt.imshow(sobelx, cmap='gray')
plt.title("4a. Sobelx ")
plt.axis('off')


plt.subplot(1 , 2 , 2)
plt.imshow(sobely, cmap='gray')
plt.title("4b. Sobely ")
plt.axis('off')
plt.show()

```

```python
sobel = cv2.magnitude(sobelx , sobely)
plt.imshow(sobel, cmap='gray')
plt.title("5. Sobel_Image")
plt.axis('off')
```
