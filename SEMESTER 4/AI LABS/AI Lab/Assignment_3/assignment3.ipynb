```python
from collections import deque
import time
import tracemalloc

mem = []
tim = []
```

```python
puzzle =  [[0, 1, 3],
            [4, 2, 6],
            [7, 5, 8]]
```

```python
end = [
    [1,2,3],
    [4,5,6],
    [7,8,0]
]
```

```python
def swap(matrix, pos1, pos2):
        matrix_copy = [row.copy() for row in matrix]
        matrix_copy[pos1[0]][pos1[1]], matrix_copy[pos2[0]][pos2[1]] = matrix_copy[pos2[0]][pos2[1]], matrix_copy[pos1[0]][pos1[1]]
        return matrix_copy

def heuristic(grid):
    c=0
    for i in range(0,3):
        for j in range(0,3):
            if(grid[i][j]==end[i][j]):
                c+=1
    return c
```

```python
def bfs(start, start0, end):
    queue = deque([(start, start0, [])])
    visited = set()

    while queue:
        current, current0, path = queue.popleft()
        x, y = current0
        if current == end:
            return path + [current]


        if 0 <= x < len(puzzle) and 0 <= y < len(puzzle[0]) and tuple(map(tuple, current)) not in visited:
            visited.add(tuple(map(tuple, current)))

            for dx, dy in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
                if 0 <= x + dx < len(puzzle) and 0 <= y + dy < len(puzzle[0]):

                    new_matrix = swap(current, (x + dx, y + dy), (x, y))
                    queue.append((new_matrix, (x + dx, y + dy), path + [current]))
```

```python
ts1 = time.time()
tracemalloc.start()

start = puzzle
start0 = (0, 0)
result = bfs(start, start0, end)
for i in range(len(result)):
  for j in range(len(result[0])):
    print(result[i][j])
  print("\n")

ts2 = time.time()
tim.append([ts2 - ts1,"BFS"])

print("Time taken by BFS: ", ts2 - ts1 )
memory = tracemalloc.get_traced_memory()
print("Memory used by BFS: ", memory[1] - memory[0])
mem.append([memory[1] - memory[0],"BFS"])
tracemalloc.stop()
```

```python
def dfs(start, start0, end):
    stack = [(start, start0, [])]
    visited = set()
    count = 0

    while stack:
        current, current0, path = stack.pop()
        x, y = current0
        if current == end:
            return path + [current]

        if 0 <= x < len(puzzle) and 0 <= y < len(puzzle[0]) and tuple(map(tuple, current)) not in visited :
            visited.add(tuple(map(tuple, current)))

            for dx, dy in [(1, 0), (0, 1), (-1, 0), (0, -1)]:
                if 0 <= x + dx < len(puzzle) and 0 <= y + dy < len(puzzle[0]):
                    new_matrix = swap(current, (x + dx, y + dy), (x, y))
                    if(tuple(map(tuple, new_matrix)) not in visited):
                        if(heuristic(new_matrix)<=heuristic(current)):
                            stack.append((new_matrix, (x + dx, y + dy), path + [current]))
```

```python
def iddfs(start, start0, end):
    depth_limit = 0
    while True:
        result = ddfs(start, start0, end, depth_limit)
        if result is not None:
            return result
        if depth_limit == len(puzzle) * len(puzzle[0]):  # Maximum depth limit
            break
        depth_limit += 1

def ddfs(start, start0, end, depth_limit):
    stack = [(start, start0, [])]
    visited = set()
    start_tuple = tuple(map(tuple, start))  # Convert the matrix to a tuple once
    visited.add(start_tuple)

    while stack:
        current, current0, path = stack.pop()
        x, y = current0
        if current == end:
            return path + [current]

        if len(path) <= depth_limit:
            for dx, dy in [(1, 0), (0, 1), (-1, 0), (0, -1)]:  # Correct movement directions
                new_x, new_y = x + dx, y + dy
                if 0 <= new_x < len(puzzle) and 0 <= new_y < len(puzzle[0]):
                    new_matrix = swap(current, (new_x, new_y), (x, y))
                    new_tuple = tuple(map(tuple, new_matrix))
                    if new_tuple not in visited:
                        visited.add(new_tuple)
                        stack.append((new_matrix, (new_x, new_y), path + [current]))

    return None
```

```python
ts1 = time.time()
tracemalloc.start()

start = puzzle
start0 = (0, 0)
result = iddfs(start, start0, end)
print("IDDFS")
for i in range(len(result)):
  for j in range(len(result[0])):
    print(result[i][j])
  print("\n")

ts2 = time.time()
print("Time taken by IDDFS: ", ts2 - ts1 )
tim.append([ts2 - ts1,"IDDFS"])

memory = tracemalloc.get_traced_memory()
print("Memory used by IDDFS: ", memory[1] - memory[0])
mem.append([memory[1] - memory[0],"IDDFS"])
tracemalloc.stop()
```
