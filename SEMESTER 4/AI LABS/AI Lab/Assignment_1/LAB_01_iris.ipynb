#**Import the DataSet**


```python
from sklearn import datasets
iris  = datasets.load_iris()
print(iris)
```

#Creating a DataFrame

```python
import pandas as pd

df = pd.DataFrame( iris.data , columns = iris.feature_names )
print (df)
```

```python
print(iris.target)
```

```python
df['label'] = iris.target
print(df.head())

```
