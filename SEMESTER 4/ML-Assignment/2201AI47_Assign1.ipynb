## Splitting the dataset into Train ,Test and Validation Set

```python
from sklearn.model_selection import train_test_split
# Splitting the data into test and train in 70 : 30
x_train , x_test, y_train , y_test = train_test_split(x,y , test_size=0.3 , random_state=0 )
```

```python
# now splitting test set into test and validation set equally 
x_val ,x_test ,  y_val, y_test  = train_test_split(x_test , y_test , test_size=0.5, random_state=0)
```

```python
## Just checking the size of training dataset 
print ("Length of training set is", len(x_train))
print ("Length of validation set is", len(x_val))
print ("Length of testing set is", len(x_test))
```

## Training the Simple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

## Predicting the Test Set Results

```python

y_pred= regressor.predict(x_test)
y_pred_val= regressor.predict(x_val)
```

## Analysing the Overfit and Underfit


```python
from sklearn.metrics import mean_squared_error, r2_score 
mean_sq_error = mean_squared_error(y_val,y_pred_val)
rootmean_sq_error = np.sqrt(mean_sq_error)
print("Validation Set Results : ")
print("MSE : ",mean_sq_error)
print("RMSE : ",rootmean_sq_error)

print("Coefficient Of Determination : ",r2_score(y_val,y_pred_val))
```

```python
y_pred_train= regressor.predict(x_train)
rmse_trainset = np.sqrt(mean_squared_error(y_train,y_pred_train))
print(f'Error of : \nValidation data : {rootmean_sq_error} \nTraining data : {rmse_trainset}')
```

```python
## Difference error in training set and validation set is comparitively very small 
## Therefore there is no overfitting in the data 
```

## Final Prediction Of Data from the Model

```python
y_test_pred = regressor.predict(x_test)

## Comparing predicted and actual values on test data set
print ("TEST SET RESULTS : ")

mean_sq_error = mean_squared_error(y_test,y_test_pred)
rootmean_sq_error = np.sqrt(rootmean_sq_error)
print("MSE =",mean_sq_error)
print("RMSE =",rootmean_sq_error)
print("Coefficient Of Determination =",r2_score(y_val,y_pred_val))
```
