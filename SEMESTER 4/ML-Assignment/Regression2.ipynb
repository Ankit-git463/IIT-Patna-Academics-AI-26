# Simple Linear Regression

## Importing Libraries

```python
import numpy as np
import pandas as pd ; 
import matplotlib.pyplot as plt 
```

## Importing the dataset 

```python
# data = pd.read_csv("AI1_20_data.csv")
# data = pd.read_csv("AI36_55_data.csv")
data = pd.read_csv("AI21_35_data.csv")
```

## Printing the data


```python
data.head()
```

```python
x=data.iloc[:,:-1].values 
y=data.iloc[:,-1].values 
```

```python
print(x)
```

```python
print(y)
```

## Visualising the Training set results

```python
plt.scatter(x,y, marker='.' )
plt.title("DATA POINTS")
plt.xlabel("Value of x")
plt.ylabel("Value of y")
plt.grid()
plt.show()

```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split

x_train , x_test, y_train , y_test = train_test_split(x,y , test_size=0.1 , random_state=0 )
```

```python
# print(x_train)
print ("Length of training dataset is", len(x_train))
```

## Training the Simple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

## Predicting the Test Set Results

```python

y_pred= regressor.predict(x_test)
```

## Plotting the training data


```python
plt.scatter(x_train,y_train, marker='.' )
plt.plot(x_train , regressor.predict(x_train) , color= 'violet')
plt.title("DATA POINTS")
plt.xlabel("Value of x")
plt.ylabel("Value of y")
plt.grid()
plt.show()
```

```python
plt.scatter(x_test,y_test , marker='.' , color='red')
# plt.plot(x_train , regressor.predict(x_train) , color= 'violet')
plt.title("Predicted Values")
plt.xlabel("Value of x")
plt.ylabel("Value of y")
plt.grid()
plt.show()
```

```python
plt.scatter(x_test,y_pred, marker='.' )
# plt.plot(x_train , regressor.predict(x_train) , color= 'violet')
plt.title("Predicted Values")
plt.xlabel("Value of x")
plt.ylabel("Value of y")
plt.grid()
plt.show()
```

```python
print("Regression Coefficient is %.2f" %regressor.coef_[0])
```

```python
# y = mx + c 

m= regressor.coef_
c_data =regressor.intercept_

print("Equation of regression line is -\n y= %.2f x" % m[0] , "+ (%.2f)" %c_data)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))


# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(y_test,y_pred))
```
